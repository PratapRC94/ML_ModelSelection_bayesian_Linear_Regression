```python
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_data = []
with open('train-crime.csv', 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
    
    
    # extracting each data row one by one
    for row in csvreader:
        train_data.append(row)

train_label = []
with open('trainR-crime.csv', 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
    
    
    # extracting each data row one by one
    for row in csvreader:
        train_label.append(row)
        
number_of_features = len(train_data[1])
number_of_examples = len(train_data)

train_data_array = np.array(train_data,dtype=float)
train_label_array = np.array(train_label,dtype=float)

test_data = []
with open('test-crime.csv', 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
    
    
    # extracting each data row one by one
    for row in csvreader:
        test_data.append(row)

test_label = []
with open('testR-crime.csv', 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
    
    
    # extracting each data row one by one
    for row in csvreader:
        test_label.append(row)
        

test_data_array = np.array(test_data,dtype=float)
test_label_array = np.array(test_label,dtype=float)


###########################TASK 2#################
identity_matrix = np.identity(number_of_features)
print(number_of_features)
phi = np.array(train_data,dtype=float)
t = np.array(train_label,dtype=float)
```

```python
alpha = 1
beta = 1.0002

while abs(alpha-beta) > 0.0001 :
    sn_inverse = alpha*identity_matrix + beta* np.dot(phi.T,phi)
    mn = beta*np.dot(np.dot(np.linalg.inv(sn_inverse),phi.T),t)
    #mn = beta*(np.dot(np.dot(np.dot(np.dot(np.linalg.inv(sn_inverse),sn_inverse),np.linalg.inv(sn_inverse)),phi.T),t))
    eigenvalues, eigenvectors = np.linalg.eig(beta*np.dot(phi.T,phi))

    gamma = 0
    for i in range(len(eigenvalues)):
        gamma = gamma + (eigenvalues[i]/(alpha + eigenvalues[i]))
    
    alpha = gamma/float((np.dot(mn.T,mn)))
    alpha = float(alpha)
    
    sum_squared_error = 0
    for i in range(len(t)):
            sum_squared_error = sum_squared_error + (t[i] - np.dot(mn.T,phi[i]))**2
    #print(sum_squared_error)

    one_by_beta = sum_squared_error/(len(t)-gamma)

    beta = 1/one_by_beta
    beta = float(beta)
    print(gamma,alpha,beta,(np.dot(mn.T,mn)))
    
print(alpha,beta)
```
